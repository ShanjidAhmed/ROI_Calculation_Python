```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Create a DataFrame from the provided dataset
data = {
    'Year': range(1, 11),  # Years 1 to 10
    'Revenue': [random.randint(100000, 200000) for _ in range(10)],  # Random revenue between $100,000 and $200,000
    'Expenses': [random.randint(60000, 120000) for _ in range(10)],  # Random expenses between $60,000 and $120,000
    'Investment': [random.randint(5000, 30000) for _ in range(10)],  # Random investment between $5,000 and $30,000
}
df = pd.DataFrame(data)

# Calculate Profit and ROI
df['Profit'] = df['Revenue'] - df['Expenses'] - df['Investment']
df['ROI'] = (df['Profit'] / df['Investment']) * 100
df['ROI'] = df['ROI'].round(2)

total_profit = df['Profit'].sum()
total_investment = df['Investment'].sum()
total_roi = (total_profit / total_investment) * 100


print("Yearly ROI:")
print(df[['Year', 'ROI']])


```

```python
# Create a line chart for ROI
plt.figure(figsize=(6, 3))
plt.plot(df['Year'], df['ROI'], marker='o', linestyle='-', color='b')
plt.title('ROI Over 5 Years')
plt.xlabel('Year')
plt.ylabel('ROI (%)')
plt.grid(True)

# Display the line chart
plt.show()

# ROI for the 5-year period
print("\nTotal ROI for 5 years:", total_roi)
```
